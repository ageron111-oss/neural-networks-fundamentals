# NumPy Expression

Using NumPy, write a function that calculates the 
following expression:

$$\exp(A^T(B + 2C) + 3I) \mathbf x,$$

where $I$ is an identity matrix of the necessary shape.

```python
from numpy.linalg import expm
def numpy_expression(A, B, C, x):
    B_plus_2C = B + 2 * C
    
    
    AT = A.T
    M = AT @ B_plus_2C
    
    n = M.shape[0]  
    I = np.eye(n)
    M_plus_3I = M + 3 * I
    
    exp_M = expm(M_plus_3I)
    
    res = exp_M @ x
    return res
```

# Einstein's Rule

In *Tensor Algebra*, a direct generalization of linear algebra to $N$-dimensional arrays called *tensors*, Einstein notation is often used.

It works as follows: if you see a repeated upper and lower index in a formula, that index is summed over.

For example, the following tensor expression, summation and matrix product are equivalent:

$$a_k^l b_l^m = \sum_{l=1}^L a_k^l b_l^m = AB$$

In this notation, a subscript means a row index and a superscript means a column index.

# Diagonal Matrix Product

You are given two square matrices: $A$ and $D$, where $A$ is a 
full matrix and $D$ is a diagonal matrix:

$$
A = \begin{bmatrix}
- & \mathbf a_1 & - \\
& \vdots & \\
- & \mathbf a_N & - \\
\end{bmatrix}
$$

$$
D = \textrm{diag}(d_1, d_2, \dots, d_N) = \begin{bmatrix}
d_1 & & & & \\
& d_2 & & & \\
& & d_3 & & \\
& & & \ddots & \\
& & & & d_N 
\end{bmatrix}
$$

Write a program to calculate $DA$ and $AD$ in the fastest possible way.

```python
def diag_prod_DA(A, D):
    d = np.diag(D)
    res =  A * d[:, np.newaxis]
    ## YOUR CODE HERE
    return res

def diag_prod_AD(A, D):
    d = np.diag(D)
    res = A * d[np.newaxis, :]
    ## YOUR CODE HERE
    return res
```

```python
answer['task4_1'] = []
answer['task4_2'] = []
for one_input in inputs['task4']:
    answer['task4_1'].append(diag_prod_DA(**one_input))
    answer['task4_2'].append(diag_prod_AD(**one_input))
```

# Sparse Matrix Product

You are given two matrices of the same shape: $A$ and $B$. Matrix $A$ is full
and is given in the form of `numpy.ndarray`.

The second matrix $B$ is **sparse**. That means that most of its entries are equal to $0$, except for $M$ nonzero entries. This matrix is given
as a set of nonzero elements in the form of a $3 \times M$ `numpy.ndarray` of row-column-value tuples (COO sparse matrix form):

$$
\begin{bmatrix}
r_1 & c_1 & v_1 \\
r_2 & c_2 & v_2 \\
& \vdots & \\
r_M & c_M & v_M \\
\end{bmatrix}
$$

If two items in this structure correspond to the same location, use the latter one.

Write the most efficient program that calculates $AB$.

Also return the ratio between the number of multiplication operations that are needed to calculate the sparse product and the number of operations for the full product.

```python
def sparse_matrix_product(A, B_sparse):
    rows = B_sparse[0, :].astype(int)   # First row: row indices
    cols = B_sparse[1, :].astype(int)   # Second row: column indices
    vals = B_sparse[2, :]               # Third row: values
    
    N, K = A.shape
    P = int(cols.max()) + 1  # Max column index + 1
    
    # Initialize result matrix
    C = np.zeros((N, P))
    
    # Count operations
    operations_sparse = 0
    M = B_sparse.shape[1]  # Number of non-zero elements
    
    # For each non-zero element in B
    for i in range(M):
        r = rows[i]
        c = cols[i]
        v = vals[i]
        
        # Add v * A[:, r] to column c of result
        C[:, c] += A[:, r] * v
        operations_sparse += N  # N multiplications per non-zero
    
    # Full matrix multiplication operations: N * K * P
    operations_full = N * K * P
    ratio = operations_sparse / operations_full
    res = C
    

    ## YOUR CODE HERE

    return res, ratio
```
